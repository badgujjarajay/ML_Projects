# Univariate Selection

```python
import pandas as pd 
import numpy as np 
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
data = pd.read_csv('./Dataset/train.csv')
```

```python
data.head()
```

```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]
```

# Apply SelectKBest

```python
bestFeatures = SelectKBest(score_func = chi2, k = 10)
```

```python
fit = bestFeatures.fit(x, y)
```

```python
fit.scores_
```

```python
dfScores = pd.DataFrame(fit.scores_)
dfColums = pd.DataFrame(x.columns)
```

```python
featureScores = pd.concat([dfColums, dfScores], axis= 1)
```

```python
featureScores.columns = ['Feature', 'Score']
```

```python
featureScores.sort_values(by='Score', ascending=False)
```

Now, we can select features which has more relevance.

# Feature Importance

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt 
```

```python
model = RandomForestClassifier()
```

```python
model.fit(x, y)
```

```python
model.feature_importances_
```

```python
featureImportance = pd.DataFrame(model.feature_importances_, index = x.columns, columns=['Importance'])
```

```python
featureImportance.sort_values(by = 'Importance', ascending=False)
```
